```python
his_ds=model.his_dataset()
```

```python
fig,axs=plt.subplots(2,1,sharex=True)

axs[0].plot(his_ds.time, -his_ds.cross_section_discharge.sel(cross_section='n_complex_xs'),
            label='n_complex_xs')

axs[0].plot(his_ds.time, his_ds.gategen_discharge.sel(gategens='pch_gate'),
           label='pch gate')

axs[0].legend()

for stn in ['pch_up','pch_down']:
    axs[1].plot(his_ds.time, his_ds.waterlevel.sel(stations=stn),label=stn)
axs[1].legend()
    
axs[1].axhline(2.10,color='k',label='Culvert invert')
```
