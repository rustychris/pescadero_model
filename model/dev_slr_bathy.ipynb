```python
from stompy.grid import unstructured_grid
from stompy.spatial import wkb2shp
import numpy as np
import matplotlib.pyplot as plt
from stompy import utils
from stompy.plot import plot_wkb
%matplotlib notebook
```

```python
g=unstructured_grid.UnstructuredGrid.read_ugrid("../grids/pesca_butano_v04/pesca_butano_asbuilt_deep_bathy.nc")
```

```python
scen_shp=wkb2shp.shp2geom("../../bathy/dem-scenarios.shp")
```

```python
poly=scen_shp['geom'][ scen_shp['src_name']=='slr_region'][0]
```

```python
dists=g.distance_transform_nodes(g.select_nodes_intersecting(poly),15.0)
```

```python
fig,ax=plt.subplots()
ax.set_adjustable('datalim')
g.plot_edges()
plot_wkb.plot_wkb(poly,alpha=0.5,color='orange')
g.plot_nodes(values=dists,clim=[0,15],zorder=3)
ax.axis((551881., 552350., 4124403., 4124985.))
```
